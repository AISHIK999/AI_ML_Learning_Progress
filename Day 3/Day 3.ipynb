## Importing the libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
```

## Loading the dataset

```python
iris = load_iris()  # load the iris dataset that we imported above
print(iris)
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)  # create a dataframe of the data. Through this we can see the features of the dataset
print(data)
```

```python
data['species'] = pd.DataFrame(iris.target) # create a dataframe of the target. Through this we can see the target of the dataset
print(data)
```

```python
data.head()
```

## Work with the dataset using decision tree and find the accuracy

```python
x = data.iloc[:, :-1]  # create a dataframe of the features. Through this we can see the features of the dataset
y = data.iloc[:, -1]  # create a dataframe of the target. Through this we can see the target of the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)  # split the data into training and testing data
model = DecisionTreeClassifier(criterion="entropy", splitter="best")    # create a decision tree classifier
model.fit(x_train, y_train)  # fit the model with the training data
y_pred = model.predict(x_test)  # predict the target of the testing data. Through this we can see the accuracy of the model
print(accuracy_score(y_test, y_pred)*100)  # print the accuracy of the model
```

```python
%matplotlib inline
# The above command is used to show the graph in the same page. Through this we can see the graph of the model we created
plt.figure(figsize=(25, 20))  # this is used to set the size of the graph. The numbers are the width and height of the graph
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)  #This is used to plot the graph of the model. The parameters depict the features, the target and the graph
plt.show() # Show the graph
```
