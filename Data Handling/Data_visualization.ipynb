# DATA VISUALIZATION
Data visualization is used to visually analyze our data
There are four ways to visually represent datasets:
- Comparison visualization is used to compare between two or more items
- Relationship visualization is used to correlate two or more items, how they work with respect to each other e.g., a scatter plot
- Distribution visualization works with the statistical distribution of a feature e.g., a histogram
- Component visualization is used to visualize the components of a dataset e.g., a stacked bar chart, pie chart

We will use the matplotlib package to visually represent our dataset

`%matplotlib [backend]`

matplotlib provides the backends: inline, notebook, qt, qt4, qt5, tk, osx, pdf, etc.

```python
import pandas as pd
%matplotlib inline
```

```python
artists = pd.read_csv("data/songs_normalize.csv")
artists.head()
```

```python
# Relationship Visualization
artists.plot(kind="scatter", x="year", y="popularity")
# Here we have plotted a scatter graph between a song's release year, and it's popularity
# We can see that 2013 marked the release of the most popular song
```

```python
# Distribution Visualization
# We can draw a histogram for a variable in our dataset (column) by passing the argument to our dataset and plotting the histogram
artists["year"].plot(kind="hist")
# From the histogram, we can see that almost 300 songs were released on 2010, and on an average 200 songs are released every year
```

```python
# Comparison Visualization
# We will draw a box plot of a continuous feature against a categorical feature, i.e., columns vs values
# vehichles.pivot(columns="categorical-component", values="continuous-component").plot(kind="box", figsize=(<diagram-width-in-inches>,<diagram-height-in-inches>))
artists.pivot(columns="year", values="popularity").plot(kind="box", figsize=(15,6))
# From the box graph, we can see that, on an average, the year 2020 didn't release much popular music
```

```python
# Composition Visualization
# We will use stacked bar charts to visualize the component makeup of data
# unstack allows us to keep the data collected from "value_counts" method and plot with the NULL values still present. By setting the stacked parameter ot True, we created a stacked bar chart
# <dataset>.groupby("column-along-x-axis")["value-along-y-axis"].value_counts().unstack().plot(kind="bar", stacked=True, figsize=(<diagram-width-in-inches>,<diagram-height-in-inches>))
artists.groupby("year")["key"].value_counts().unstack().plot(kind="bar", stacked=True, figsize=(10,6))
```
